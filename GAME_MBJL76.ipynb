# PART 1  Medialib mini game

###### Step 1: Draw the Initial Game Background
The initial background of the game is represented by the list background1. Each character in the list represents an image ('X' for walls, '_' for paths, 'M' for monsters, and specific letters for airline logos). The avatar ('H') starts at a specific position.

```python
from medialib import *

initialize()

# Initial game setup
background1 = [
    ['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X'],
    ['X','H','_','L','_','_','M','_','_','E','_','_','J','_','_','_','_','X'],
    ['X','_','_','_','M','_','B','_','L','_','_','_','_','B','_','J','_','X'],
    ['X','_','_','B','_','_','_','_','_','J','_','_','E','_','L','_','_','X'],
    ['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X']
]

no_of_rows_in_array = len(background1)
no_of_columns_in_array = len(background1[0])

position_array = [
    [['0' for col in range(3)] for col in range(no_of_columns_in_array)] for row in range(no_of_rows_in_array)
]

x0 = 10
y0 = 200
image_row = 0
image_column = 0

game_running = True
CurrentScore = 0
end_game_over = False
illegal_input_britishAirways = False

size = 30
startX = 0
startY = 0
NstartX = 40
NstartY = 10
```

###### Step 2: Avatar Movement Controls
The player controls the avatar using keys:

'a' for left
'd' for right
'w' for up
's' for down
'q' to quit the game
The avatar cannot pass through walls ('X'). If it encounters a monster ('M'), the game ends with a message.

###### Step 3: Handling Airplane Logo Interactions
When the avatar moves onto a position with an airline logo ('B', 'E', 'L', 'J'), it replaces the logo with 'H' (avatar) and displays a message. Points increase by 1. When the avatar leaves, the logo is replaced back with '_' (road).

###### Step 3: Handling Airplane Logo Interactions
When the avatar moves onto a position with an airline logo ('B', 'E', 'L', 'J'), it replaces the logo with 'H' (avatar) and displays a message. Points increase by 1. When the avatar leaves, the logo is replaced back with '_' (road).



###### Step 4: British Airways Interaction
If the avatar moves onto a position with 'B.png' (British Airways), it triggers a special interaction:

Displays a welcome message and increases points by 1.
Asks the player to make a decision ('h' or 'i').
Depending on the decision, additional points are awarded and the game state updates accordingly.


###### Step 5: Handling Illegal Input
If the player enters an invalid key, an error message is displayed, and the input is requested again.



###### Step 6: Winning Condition
If the avatar collects 20 points or more, the game displays 'you_win.png' and ends.

```python
###### from medialib import *

initialize()

def create_position_array(x, y, image_row, image_column, image):
    position_array[image_row - 1][image_column - 1] = [x, y, image]

def create_background(j, x0, y0):
    image_mapping = {
        "X": 'X.png',
        "_": '_.png',
        "L": 'L.png',
        "M": 'M.png',
        "J": 'J.png',
        "E": 'E.png',
        "B": 'B.png',
        "H": 'H.png',
    }

    image_path = image_mapping.get(j, 'default.png')
    draw(image_path, x0, y0)

def set_column(i, y0, image_row, x0, image_column):
    for j in i:
        image_column += 1
        x0 += 30 + (image_column - 2) * 0.2 if image_column > 1 else 0
        create_background(j, x0, y0)
        if game_running:
            create_position_array(x0, y0, image_row, image_column, j)

def set_row(background, image_column, y0, x0, image_row):
    for i in background:
        image_row += 1
        y0 += 30 + (image_row - 2) * 0.2 if image_row > 1 else 0
        set_column(i, y0, image_row, x0, image_column)

def calculate_current_score(score):
    global CurrentScore
    CurrentScore += score
    rect(10, 20, 200, 20, r=200, g=200, b=200)
    text(f"Current points: {CurrentScore:.2f}", 10, 20, 16)
    if CurrentScore >= 20:
        global end_game_over
        end_game_over = True
        clear(200, 200, 200)
        draw('you_win.png', 100, 50)
        display_text("You win. Please click to exit", 250, 20, 20)

def display_text(str, x, y, clr):
    text(str, x, y, clr)

def game_background():
    text('Maze game', 300, 100, 20)
    text('You are a lost hero and have to find your way back home...', 10, 500, 16)
    text('When you collect 20 points, you can get back home!', 10, 520, 16)
    text('To play: press a and d keys to move left and right', 10, 540, 16)
    text('         press w and s keys to move upwards and downwards', 10, 560, 16)
    text('         press q to quit the game', 10, 580, 16)

def british_airway():
    global CurrentScore, illegal_input_britishAirways
    if illegal_input_britishAirways == False:
        clear(200, 200, 200)
        text(f"Current points: {CurrentScore:.2f}", 10, 20, 16)
        text("Welcome on board with British Airways! You just received 1 point!", 50, 50, 16)
        draw('I.png', 10, 200)
        text('However, British Airways has an employee work crisis.', 10, 500, 16)
        text('If you help them make the decision, you can get extra points!', 10, 520, 16)
        text('   h. Hire new staff', 10, 540, 16)
        text('   i. Increase existing staff pay', 10, 560, 16)
        text('Please make your decision by inputting "h" or "i"', 10, 580, 16)

    user_input = wait_key_press()
    if user_input == 'h':
        clear(200, 200, 200)
        game_background()
        illegal_input_britishAirways = False
        set_row(background1, image_column, y0, x0, image_row)
        display_text('Thank you for supporting hiring new staff! You received extra 5 points!', 50, 50, 16)
        calculate_current_score(5)
    elif user_input == 'i':
        clear(200, 200, 200)
        game_background()
        illegal_input_britishAirways = False
        set_row(background1, image_column, y0, x0, image_row)
        display_text('Thank you for supporting an increase in staff pay! You received extra 4 points!', 50, 50, 16)
        calculate_current_score(4)
    else:
        rect(10, 50, 700, 20, r=200, g=200, b=200)
        text("please enter a valid key", 10, 50, 16)
        illegal_input_britishAirways = True
        british_airway()

def restructure_background(avatar_xpos, avatar_ypos, next_xpos, next_ypos):
    global end_game_over
    is_british_airway = False
    if position_array[next_xpos][next_ypos][2] in ["L", "J", "E"]:
        text(f"Welcome on board with {position_array[next_xpos][next_ypos][2]}! You just received 1 point!", 50, 50, 16)
        calculate_current_score(1)
    elif position_array[next_xpos][next_ypos][2] == "B":
        is_british_airway = True
        calculate_current_score(1)
        if not end_game_over:
            british_airway()
    if not end_game_over:
        background1[avatar_xpos][avatar_ypos] = "_"
        background1[next_xpos][next_ypos] = "H"
        create_position_array(
            position_array[avatar_xpos][avatar_ypos][0], position_array[avatar_xpos][avatar_ypos][1],
            avatar_xpos + 1, avatar_ypos + 1, "_"
        )
        create_position_array(
            position_array[next_xpos][next_ypos][0], position_array[next_xpos][next_ypos][1],
            next_xpos + 1, next_ypos + 1, "H"
        )
        if is_british_airway:
            set_row(background1, image_column, y0, x0, image_row)

row_no = 0
column_no = 0

def find_index(str, row_no, column_no, user_input):
    global end_game_over
    element_found = False
    for i in background1:
        for avatar in i:
            if avatar == str:
                column_no = i.index(str)
                element_found = True
                break
        if element_found:
            break
        else:
            row_no += 1
    if user_input == 'a':
        if position_array[row_no][column_no - 1][2] == "X":
            display_text("The avatar cannot go through a Wall", 10, 50, 16)
        elif position_array[row_no][column_no - 1][2] == "M":
            clear()
            text("The Avatar has come across a Monster! It has died. GAME OVER", 50, 50, 16)
            end_game_over = True
        else:
            create_background("_", position_array[row_no][column_no][0], position_array[row_no][column_no][1])
            create_background("H", position_array[row_no][column_no - 1][0], position_array[row_no][column_no - 1][1])
            restructure_background(row_no, column_no, row_no, column_no - 1)
    elif user_input == 'd':
        if position_array[row_no][column_no + 1][2] == "X":
            display_text("The avatar cannot go through a Wall", 10, 50, 16)
        elif position_array[row_no][column_no + 1][2] == "M":
            clear()
            text("The Avatar has come across a Monster! It has died. GAME OVER", 50, 50, 16)
            end_game_over = True
        else:
            create_background("_", position_array[row_no][column_no][0], position_array[row_no][column_no][1])
            create_background("H", position_array[row_no][column_no + 1][0], position_array[row_no][column_no + 1][1])
            restructure_background(row_no, column_no, row_no, column_no + 1)
    elif user_input == 'w':
        if position_array[row_no - 1][column_no][2] == "X":
            display_text("The avatar cannot go through a Wall", 10, 50, 16)
        elif position_array[row_no - 1][column_no][2] == "M":
            clear()
            text("The Avatar has come across a Monster! It has died. GAME OVER", 50, 50, 16)
            end_game_over = True
        else:
            create_background("_", position_array[row_no][column_no][0], position_array[row_no][column_no][1])
            create_background("H", position_array[row_no - 1][column_no][0], position_array[row_no - 1][column_no][1])
            restructure_background(row_no, column_no, row_no - 1, column_no)
    elif user_input == 's':
        if position_array[row_no + 1][column_no][2] == "X":
            display_text("The avatar cannot go through a Wall", 10, 50, 16)
        elif position_array[row_no + 1][column_no][2] == "M":
            clear()
            text("The Avatar has come across a Monster! It has died. GAME OVER", 50, 50, 16)
            end_game_over = True
        else:
            create_background("_", position_array[row_no][column_no][0], position_array[row_no][column_no][1])
            create_background("H", position_array[row_no + 1][column_no][0], position_array[row_no + 1][column_no][1])
            restructure_background(row_no, column_no, row_no + 1, column_no)

background1 = [
    ['X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X'],
    ['X', 'H', '_', 'L', '_', '_', 'M', '_', '_', 'E', '_', '_', 'J', '_', '_', '_', '_', 'X'],
    ['X', '_', '_', '_', 'M', '_', 'B', '_', 'L', '_', '_', '_', '_', 'B', '_', 'J', '_', 'X'],
    ['X', '_', '_', 'B', '_', '_', '_', '_', '_', 'J', '_', '_', 'E', '_', 'L', '_', '_', 'X'],
    ['X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X']
]

no_of_rows_in_array = len(background1)
no_of_columns_in_array = len(background1[0])

position_array = [
    [['0' for col in range(3)] for col in range(no_of_columns_in_array)] for row in range(no_of_rows_in_array)
]

x0 = 10
y0 = 200
image_row = 0
image_column = 0

clear(200, 200, 200)
game_background()

game_running = True
CurrentScore = 0
set_row(background1, image_column, y0, x0, image_row)
calculate_current_score(0)
end_game_over = False
illegal_input_britishAirways = False

while game_running:
    global end_game_over
    if not end_game_over:
        user_input = wait_key_press()
        if user_input in ['a', 'd', 'w', 's']:
            rect(10, 50, 700, 20, r=200, g=200, b=200)
            find_index('H', row_no, column_no, user_input)
        elif user_input == 'q':
            clear()
            display_text("You have quit the game. Please click to exit", 200, 20, 20)
            break
        else:
            rect(10, 50, 700, 20, r=200, g=200, b=200)
            display_text("please enter a valid key", 10, 50, 16)
    else:
        break

wait_mouse_leftclick()
all_done()
```
